```python
# Lista para almacenar objetos PDF
pdfs = []

# Leer cada archivo PDF en el directorio y agregarlo a la lista 'pdfs'
for archivo in archivos:
    pdfs.append(pdfrw.PdfReader(directorio + archivo))
```

```python
# Lista para almacenar URLs
urls = []

# Iterar sobre cada PDF en la lista 'pdfs' y extraer las URLs de los enlaces
for pdf in pdfs:
    for page in pdf.pages:
        for annot in page.Annots or []:
            urls.append(annot.A.URI)

# Eliminar duplicados de la lista de URLs
urls = list(set(urls))

# Utilizar expresiones regulares para limpiar las URLs y quitar caracteres no deseados
urls = [re.search(r'\((.*?)\)', elemento).group(1).strip("'") if re.search(r'\((.*?)\)', elemento) else elemento for elemento in urls]

# Eliminar una URL específica que no se desea procesar
urls.remove('https://www.ekohealth.com')

urls
```

```python
# Configurar opciones para el navegador Chrome
opts = Options()
```

```python
# Iterar sobre cada URL en la lista 'urls'
for url in urls:
    print(url)
    # Configurar el navegador Chrome y abrir la URL
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=opts)
    driver.get(url)
    # Encontrar y hacer clic en un botón específico en la página web
    button = driver.find_element(By.XPATH, '//button[@class="chakra-button css-16qccfi"]')
    button.click()
    # Esperar 2 segundos antes de continuar con la siguiente URL
    time.sleep(2)
```
